# Amazon Scrape

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from sqlalchemy import create_engine


url_str = input("Enter the product you want to search: ")
url = []

words = url_str.split()
var = len(words)


# url = "https://www.amazon.com/s/ref=nb_sb_noss_1?url=search-alias%3Daps&field-keywords=iphone%208"
```

```python
if var == 1:
    url = "https://www.amazon.com/s/ref=nb_sb_noss_1?url=search-alias%3Daps&field-keywords=" + words[0]

if var == 2:
    url = "https://www.amazon.com/s/ref=nb_sb_noss_1?url=search-alias%3Daps&field-keywords=" + words[0] + "+" + words[1]

elif var == 3:
    url = "https://www.amazon.com/s/ref=nb_sb_noss_1?url=search-alias%3Daps&field-keywords=" + words[0] + "+" + words[1] + "+" + words[2]
```

```python
# headers = {'User-Agent': 'Mozilla/5.0 (compatible; Googlebot/2.1; +http://www.google.com/bot.html'}
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/77.0.3865.120 Safari/537.36'}

response = requests.get(url, headers=headers)

soup = BeautifulSoup(response.content, "lxml")

print(soup.prettify())
```

```python
results = soup.find_all('div', class_="s-include-content-margin s-border-bottom")
results
```

```python
product=[]
price=[]
review=[]
link=[]

for page in range(1,500):

    for result in results:
        try:
            product_name = result.find('span', class_="a-size-medium a-color-base a-text-normal").text
            price_whole = result.find('span',class_='a-price-whole').text
            price_fraction = result.find('span',class_='a-price-fraction').text
            reviews = result.find('span',class_='a-icon-alt').text
            links = "https://www.amazon.com" + result.find('div', class_='a-section a-spacing-none').find('span', class_='rush-component').find('a',class_='a-link-normal')["href"]
       

            if (product_name and price_whole and price_fraction):
                product.append(product_name)
                price.append(price_whole + price_fraction)
                review.append(reviews)
                link.append(links)
          
        except AttributeError as e:
            print(e)
```

```python
amazon = {"Product_Name":product,"Price":price,"Rating":review,"Link":link}

amazon_df=pd.DataFrame(amazon)

amazon_df
```

```python
amazon_df.iloc[0,3]
```

```python
amazon_df["Product_Name"]=amazon_df["Product_Name"].str.split("iPhone 7",n=1,expand=True)
amazon_df.head()
```

```python
amazon_df["Product_Name"]=amazon_df["Product_Name"].str.split("iPhone 6",n=1,expand=True)
amazon_df.head()
```

```python
amazon_df["Product_Name"]=amazon_df["Product_Name"].str.split("iPhone 10",n=1,expand=True)
amazon_df.head()
```

```python
amazon_df=amazon_df[amazon_df['Product_Name'] != "Apple "]
amazon_df.head()
```

```python
amazon_df = amazon_df.rename(columns={"Product_Name": "product_name","Price": "price","Rating":"rating"})
```

```python
amazon_df['id']= range(1, 1+len(amazon_df))
amazon_df=amazon_df[["id","product_name","price","rating"]]
amazon_df.head()
```

```python
connection_string = "postgres:postgres@localhost:5432/product_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
amazon_df.to_sql(name='iphone_8', con=engine, if_exists='append', index=False)
```
